```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(3)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.subplot(3,2,1)
plt.plot(x, x**2+2, color ='g')
plt.subplot(3,2,2)
plt.plot(x, 3*x**3-1, color='b')
plt.subplot(3,2,3)
plt.plot(x, 1/x**2+2, color='y')
plt.subplot(3,2,4)
plt.plot(x, 2*x**3+2, color='r')
plt.show()
```
